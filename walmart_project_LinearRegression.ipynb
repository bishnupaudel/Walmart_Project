```python

# Feature identifying using VIF
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def get_vif_factors( X ):
    X_matrix = X.to_numpy()
    vif = [ variance_inflation_factor( X_matrix, i ) for i in range( X_matrix.shape[1] ) ]
    vif_factors = pd.DataFrame()
    vif_factors['column'] = X.columns
    vif_factors['vif'] = vif
    return vif_factors
```

```python
data_store_1[X_features]
```

```python
vif_factors = get_vif_factors( data_store_1[X_features] )
vif_factors
```

```python
columns_with_large_vif = vif_factors[vif_factors.vif > 4].column
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
plt.figure( figsize = (12,10) )
sn.heatmap( data_store_1[columns_with_large_vif].corr(), annot = True );
plt.title( "Figure 4.5 - Heatmap depicting correlation between features");
```
